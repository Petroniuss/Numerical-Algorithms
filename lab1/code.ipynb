```python
import matplotlib.pyplot as plt
import numpy as np
import time as t
import math

plt.rcParams['figure.dpi'] = 180

N  = 10 ** 7
v  = np.float32(0.53125)
correct_sum = np.float32(5312500.0)
xs = np.empty(N, dtype=np.float32)
xs.fill(v)

def sum(xs):
    acc = np.float32(0.0)
    for x in xs:
        acc += x
    
    return acc

def measure_time(alg, arg):
    start  = t.time()
    result = alg(arg)
    end    = t.time()
    
    return result, (end - start)

def abs_error(correct, actual):
    return abs(correct - actual)

def rel_error(correct, actual):
    return abs_error(correct, actual) / (correct)
```

```python
result, running_time = measure_time(sum, xs)
print(f"Measued sum amounts to: {result} when correct value is {correct_sum}")
print(f"Absolute error: {abs_error(correct_sum, result)}")
print(f"Relative error: {rel_error(correct_sum, result)}")
print(f"Measured time: {running_time}s")
```

```python
def measure_err(xs):
    errors = []
    running_sum = np.float32(0.0)
    correct = np.float32(0.0)
    for i, x in enumerate(xs):
        running_sum += x
        if (i + 1) % 25000 == 0:
            correct += np.float32(13281.25)
            errors.append(abs_error(correct, running_sum))
            
    return errors



y = measure_err(xs)
x = np.arange(len(y))

plt.scatter(x, y)
plt.savefig("error-1-3.png")
plt.show()
```

```python
def fancySum(xs):
    def help(ys, to):
        i = 0
        while (i + 1) < to:                
            e1 = ys[i]
            e2 = ys[i + 1]
            ys[i // 2] = e1 + e2
            
            i += 2
            
        if (i + 1) == to:
            ys[i // 2] = ys[i]
            i += 2
            
        return i // 2
    
    
    to = len(xs)
    while to != 1:
        to = help(xs, to)
        
    return xs[0]
```

```python
result, running_time = measure_time(sum, xs)

print("-" * 3 + "PLAIN-SUM" + ("-" * 3))
print(f"Measured sum amounts to: {result} when correct value is {correct_sum}")
print(f"Absolute error: {abs_error(correct_sum, result)}")
print(f"Relative error: {rel_error(correct_sum, result)}")
print(f"Measured time: {running_time}s")

ys = np.copy(xs)
result, running_time = measure_time(fancySum, ys)

print("-" * 3 + "RECURSISVE-SUM" + ("-" * 3))
print(f"Measured sum amounts to: {result} when correct value is {correct_sum}")
print(f"Absolute error: {abs_error(correct_sum, result)}")
print(f"Relative error: {rel_error(correct_sum, result)}")
print(f"Measured time: {running_time}s")
```

```python
### Sorted numbers proove this algorithm to be inefficient since it leads to adding numbers which vary a lot.
def edge_case():
    return np.array([np.float32(i) for i in range(10000)], dtype = np.float32)

correct_sum_edge_case = np.float32(50005000)
result, running_time = measure_time(fancySum, edge_case())

print("-" * 3 + "RECURSISVE-SUM --> EDGE-CASE" + ("-" * 3))
print(f"Measured sum amounts to: {result} when correct value is {correct_sum_edge_case}")
print(f"Absolute error: {abs_error(correct_sum_edge_case, result)}")
print(f"Relative error: {rel_error(correct_sum_edge_case, result)}")
print(f"Measured time: {running_time}s")
```

```python
def kahan_sum(xs):
    acc = np.float32(0.0)
    err = np.float32(0.0)
    
    for x in xs:
        y = x - err
        temp = acc + y
        err = (temp - acc) - y
        acc = temp
        
    return acc
```

```python
result, running_time = measure_time(kahan_sum, xs)

print("-" * 3 + "KAHAN-SUM" + ("-" * 3))
print(f"Measured sum amounts to: {result} when correct value is {correct_sum}")
print(f"Absolute error: {abs_error(correct_sum, result)}")
print(f"Relative error: {rel_error(correct_sum, result)}")
print(f"Measured time: {running_time}s")

ys = np.copy(xs)
result, running_time = measure_time(fancySum, ys)

print("-" * 3 + "RECURSISVE-SUM" + ("-" * 3))
print(f"Measured time: {running_time}s")
```

```python
def dzeta_forward(s, n, precision = np.float32):
    acc = precision(0.0)
    for i in range(1, n + 1):
        acc += precision(1.0) / precision(i ** s))
    
    return acc

def dzeta_backwards(s, n, precision = np.float32):
    acc = precision(0.0)
    for i in range(n, 0, -1):
        acc += precision(1.0) / precision(i ** s)
            
    return acc

def eta_forward(s, n, precision = np.float32):
    acc = precision(0.0)
    for i in range(1, n + 1):
        acc += precison((-1) ** (i - 1)) / precision(i ** s)

    return acc
        
def eta_backwards(s, n, precision = np.float32):
    acc = precision(0.0)
    for i in range(n, 0, -1):
        acc += precison((-1) ** (i - 1)) / precision(i ** s)
        
    return acc
```

```python
def logisitc(r, xn):
    return (r * xn) * (1 - xn)
```

```python
r = 2.8
x, y = [], []
x0 = 1e-5
iterations = 2000
last       = 400

while r <= 4:
    r += .005
    xn = x0
    for i in range(iterations):
        xn = logisitc(r, xn)
        if i >= (iterations - last):
            x.append(r)
            y.append(xn)
    
        

plt.scatter(np.array(x), np.array(y), s = 0.1, alpha = .2, color = 'k')
plt.legend(['Bifurcation diagram'], loc='upper left')
# plt.savefig('bifur1.svg', bbox_inches='tight', format='svg', dpi=1200)
plt.savefig('bifur1.png', bbox_inches='tight')
plt.show()
```
