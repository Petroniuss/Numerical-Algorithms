```python
import numpy as np
from numpy.fft import fft, ifft

import matplotlib.pyplot as plt
from matplotlib import cm

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16,9)
```

```python
t = np.arange(0, .25, .001)
n = len(t)

f1 = np.sin(2 * np.pi * 20 * t)
f2 = np.sin(2 * np.pi * 60 * t)
f3 = np.sin(2 * np.pi * 120 * t)

m = n // 3
f1p = f1.copy()
f1p[m:] = 0

f2p = f2.copy()
f2p[:m] = 0
f2p[2*m:] = 0

f3p = f3.copy()
f3p[:2*m] = 0

sig1 = f1 + f2 + f3
sig2 = f1p + f2p + f3p
```

## First signal in time domain

```python
plt.plot(t, sig1, color = 'black')
```

## Second signal in time domaing

```python
plt.plot(t, sig2, color = 'green')
```

## First signal in frequency domain

```python
freq = np.arange(n) / (.001 * n)
y    = fft(sig1)
plt.plot(freq, np.absolute(y))
```

```python
plt.plot(freq, np.angle(y))
```

## Second signal in frequency domain

```python
freq = np.arange(n) / (.001 * n)
y    = fft(sig2)
plt.plot(freq, np.absolute(y))
```

```python
plt.plot(freq, np.angle(y))
```

## Noise filtering 

out of curiosity..

```python
from numpy.fft import fft, ifft

t = np.arange(0, .25, .001)
n = len(t)

x = np.sin(2 * np.pi * 20 * t) + np.sin(2 * np.pi * 120 * t)
y = x + np.random.randn(n)

plt.plot(t, x, color = 'black')
plt.plot(t, y, color = 'green')
```

```python
Y = fft(y)
freq = np.arange(n) / (.001 * n)
plt.plot(freq, np.absolute(Y))
```

```python
S = Y.copy()
S[np.absolute(S) < 50] = 0
y = ifft(Y)
s = ifft(S)

plt.plot(t, y, color = 'green')
plt.plot(t, s, color = 'black')
```
